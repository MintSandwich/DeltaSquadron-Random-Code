```python
import matplotlib.pyplot as plt

import tensorflow as tf
import  tensorflow_datasets as tsdf
import tensorflow_addons as tfa
```

```python
data = tsdf.load(name='horses_or_humans', split='train', as_supervised = True)
val_data = tsdf.load(name='horses_or_humans', split='test', as_supervised = True)


```

```python
train_batches = data.shuffle(100).batch(32)
validation_batches = val_data.batch(32)
```

```python
train_batches = data.shuffle(100).batch(32)
validation_batches = val_data.batch(32)

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu',
                                      input_shape=(300, 300, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])


model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_batches, epochs=10, validation_data=validation_batches, validation_steps=1)
```
